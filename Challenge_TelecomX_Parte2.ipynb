#Encoding

Transforma las variables categóricas a formato numérico para hacerlas compatibles con los algoritmos de machine learning.


```python
# Separación de variables

#Variables explicativas
X=datos_encoding.drop('Cancelacion',axis=1)

#Variable de respuesta
y=datos_encoding['Cancelacion']
```

```python
columnas= X.columns
```

```python
one_hot= make_column_transformer((OneHotEncoder(drop='if_binary'),
                                  ['Genero', 'Adulto_Mayor', 'Asociado', 'Dependientes','Servicio_Telefonico', 'Lineas_Multiples',
                                   'Servicio_Internet','Seguridad_En_Linea', 'Respaldo_En_Linea', 'Proteccion_Dispositivo','Soporte_Tecnico',
                                   'TV_Streaming', 'Peliculas_Streaming', 'Contrato', 'Facturacion_Electronica', 'Metodo_Pago']),
                                 remainder='passthrough', sparse_threshold=0, force_int_remainder_cols=False)
```

```python
X=one_hot.fit_transform(X)
```

```python
one_hot.get_feature_names_out(columnas)
```

```python
pd.DataFrame(X,columns=one_hot.get_feature_names_out(columnas))
```

#Separación de datos

Divide el conjunto de datos en entrenamiento y prueba para evaluar el rendimiento del modelo. Una división común es 70% para entrenamiento y 30% para prueba, o 80/20, dependiendo del tamaño de la base de datos.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=5)
```

#Creación de modelos

Regresión Logística / KNN: Estos modelos son sensibles a la escala de los datos, por lo que la normalización es importante para que los coeficientes o las distancias se calculen correctamente.

Árbol de Decisión / Random Forest: Estos modelos no dependen de la escala de los datos, por lo que no es necesario aplicar normalización.

Si decides normalizar los datos, deberías explicar cómo esta etapa asegura que los modelos basados en distancia o en optimización de parámetros no se vean sesgados por la magnitud de las variables.

##Modelo dummy

```python
from sklearn.dummy import DummyClassifier
```

```python
dummy=DummyClassifier()
dummy.fit(X_train,y_train)

dummy.score(X_test,y_test)
```

##Modelo Arbol de decisión

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
modelo_arbol=DecisionTreeClassifier(max_depth=3, random_state=5)
modelo_arbol.fit(X_train,y_train)

modelo_arbol.score(X_test,y_test)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
nombre_columnas=one_hot.get_feature_names_out(columnas)
nombre_columnas
```

```python
plt.figure(figsize=(30,20))
plot_tree(modelo_arbol,filled=True, class_names=['No','Sí'],fontsize=6, feature_names=nombre_columnas,);
```

```python
modelo_arbol.score(X_train,y_train)
```

##Modelo KNN

Normalización

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
normalizacion= MinMaxScaler()
X_train_normalizado=normalizacion.fit_transform(X_train)
```

Algoritmo KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_test_normalizado=normalizacion.transform(X_test)
```

```python
knn=KNeighborsClassifier()
knn.fit(X_train_normalizado,y_train)

knn.score(X_test_normalizado,y_test)
```

Exactitud preliminar

```python
#Exactitud sin X_validacion y y_validacion
lista=[('dummy', dummy,X_test),('de arbol', modelo_arbol,X_test),('knn', knn,X_test_normalizado)]
for i in lista:
    print(f'La exactitud del modelo {i[0]}: {i[1].score(i[2],y_test)}')
```

#Evaluación de modelos

```python
#Creando un conjunto de validación
X_train, X_val, y_train, y_val=train_test_split(X,y,stratify=y, random_state=5)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score
```

##Modelo de arbol

###Matriz de confusión


```python
from sklearn.metrics import confusion_matrix
```

```python
y_previsto=modelo_arbol.predict(X_val)
matriz_confusión=confusion_matrix(y_val,y_previsto)
print(matriz_confusión)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
visualizacion=ConfusionMatrixDisplay(confusion_matrix=matriz_confusión,display_labels=['No Canceló','Si Canceló'])
visualizacion.plot()
```

###Exactitud, precision, recall, F1-score

```python
print(f'La presicion del modelo es: {precision_score(y_val,y_previsto)}')
print(f'La sensibilidad(recall) del modelo es: {recall_score(y_val,y_previsto)}')
print(f'La exactitud(accury) del modelo es: {accuracy_score(y_val,y_previsto)}')
print(f'La f1-score del modelo es: {f1_score(y_val,y_previsto)}')
```

##Modelo KNN

###Matriz de confusión

```python
X_val_normalizado=normalizacion.transform(X_val)
```

```python
y_previsto_knn=knn.predict(X_val_normalizado)
matriz_confusión=confusion_matrix(y_val,y_previsto)
print(matriz_confusión)
```

```python
visualizacion=ConfusionMatrixDisplay(confusion_matrix=matriz_confusión,display_labels=['No Canceló','Si Canceló'])
visualizacion.plot()
```

###Exactitud, precision, recall, F1-score

```python
print(f'La presicion del modelo es: {precision_score(y_val,y_previsto_knn)}')
print(f'La sensibilidad(recall) del modelo es: {recall_score(y_val,y_previsto_knn)}')
print(f'La exactitud(accury) del modelo es: {accuracy_score(y_val,y_previsto_knn)}')
print(f'La f1-score del modelo es: {f1_score(y_val,y_previsto_knn)}')
```

##Comportamiento del modelo con los coeficientes de correlación

El principal factore que genera mayor correlación en <font color="red">**Cancelación**</font> se presentan en la sección de <font color="blue">**Análisis de correlación**</font>, siendo representado el tipo de contrato <font color="green">**Mes a mes**</font> como  aquel con un mayor coeficiente, contando con una tendencia de correlación positiva, lo que quiere decir que a medida que una variable aumenta, la otra también aumenta.

##Métrica de evaluación

Para la evaluación del desempeño de los modelos de machine learning nos enfocaremos en el uso del métrico de evaluación  **F1-score**, presentado en la sección <font color="blue">**Evaluación de modelos**</font> ya que nos ayuda con el equilibrio entre el recall y la precisión, siendo útil cuando las clases de la variable objetivo <font color="red">**(Cancelación)**</font> están desbalanceadas, es decir, cuando hay una cantidad de datos diferente para cada clase <font color="green">**' No canceló '**</font> y <font color="green">**' Si canceló '**</font>, este método queda como soporte de las técnicas de oversamplig o undersamplig (estrategias para igualar la cantidad de datos que la clase de mayor o menor frecuencia).

##Mejor opción para modelo de machine learning
Para este proyecto el modelo de machine learning mejor adaptado es el <font color="blue">**Modelo KNN**</font> ya que lleva una mejor aproximación para los datos previsto dentro de la matriz de confusión. La regla de evaluación para F1-score expone que el coeficiente el resultado ideal de la operación es de 1.  Al tener un coeficiente de 0.50, mayor que en el <font color="blue">**Modelo de árbol**</font> que fue de 0.44, tomamos como modelo objetivo el de **KNN**, dejando el resultado de la métrica en una proporción del 50%, dismuniyendo así el factor de riesgo de falsos positivos-negativos.
